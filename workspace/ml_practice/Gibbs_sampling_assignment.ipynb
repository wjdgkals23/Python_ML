```r
#  

Nsim=6000
x1=numeric(Nsim)
x2=numeric(Nsim)

x1[1] <- 0.1
x2[1] <- 0.1

for (i in 2:Nsim){
  x1[i]=rexp(1,x2[i-1]+1)
  x2[i]=rexp(1,x1[i]+1)
}

ts.plot(x1,col=3, main = "Traceplot")
ts.plot(x2,col=3, main = "Traceplot")

library(MASS)
contour(kde2d(x1, x2, lims= c(-3,3,-3,3)), xlab='x1', ylab='x2',main='Estimated Kernel Density')

for (i in 1:Nsim){
  segments(x1[i],x2[i],x1[i+1],x2[i],lwd=2,col="blue")
  segments(x1[i+1],x2[i],x1[i+1],x2[i+1],lwd=2,col="blue")
}


library(MASS)
contour(kde2d(x1, x2, lims= c(-3,3,-3,3)), xlab='x1', ylab='x2',main='Estimated Kernel Density')

points(x1,x2,pch=20,col="red")
```
