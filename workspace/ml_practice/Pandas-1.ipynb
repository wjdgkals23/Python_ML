```python
titanic = sns.load_dataset("titanic")
```

```python
def get_values_count_each_row(df):
    for i in df.columns:
        print(df.loc[:,i].value_counts())
```

```python
get_values_count_each_row(titanic.sort_values(by="age"))
# titanic 데이터를 age를 기준으로 정렬한 후 (의미없는 짓거리) 
#각 열에 대해 원소들의 갯수를 파악한다.
```

```python
titanic.sort_values(by="survived")
```

```python
survived = titanic[titanic["survived"] == 1]
death = titanic[titanic["survived"] == 0]
```

```python
survived_nnan = survived[~np.isnan(survived["age"])]
death_nnan = death[~np.isnan(death["age"])]
```

```python
bins = [1,5, 25, 35, 60, 99]
labels = ["미성년자", "청년", "중년", "장년", "노년"]

death_cats = pd.cut(death_nnan["age"], bins, labels=labels)
survived_cats = pd.cut(survived_nnan["age"], bins, labels=labels)
```

```python
round(pd.value_counts(death_cats)/death_nnan["age"].count()*100,0)
```

```python
round(pd.value_counts(survived_cats)/survived_nnan["age"].count()*100,0)
```
