```python
# setting library
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
import time
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA

# setting data
from six.moves import urllib
from sklearn.datasets import fetch_mldata
mnist = fetch_mldata('MNIST original') # 데이터 로딩

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

X = mnist["data"] 
y = mnist["target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=3) 

X_train.shape, X_test.shape

rf = RandomForestClassifier(n_jobs=-1, n_estimators=100)

ts = time.time()
rf.fit(X_train, y_train)
te = time.time()
t_stamp = te-ts
t_stamp

lr = LogisticRegression(solver='lbfgs') 
# 'lbfgs'는 quasi-Newton 방식의 옵티 마이저입니다. 좋다기래 사용해봤어요!

ts1 = time.time()
lr.fit(X_train, y_train)
te1 = time.time()
t_stamp1 = te1 - ts1
t_stamp1

rf.score(X_test, y_test)

lr.score(X_test, y_test)

pca = PCA()
tp_X_train = pca.fit(X_train)

cumsum = np.cumsum(pca.explained_variance_ratio_)
d = np.argmax(cumsum >= 0.96) + 1


pca1 = PCA(n_components=d)
X_reduced = pca1.fit_transform(X_train)
X_reduced_test = pca1.fit_transform(X_test)


np.sum(pca1.explained_variance_ratio_)

rf2 = RandomForestClassifier(n_jobs=-1, n_estimators=100)

ts2 = time.time()
rf2.fit(X_reduced, y_train)
te2 = time.time()
t_stamp2 = te2-ts2
t_stamp2

lr2 = LogisticRegression(solver='lbfgs') 
ts3 = time.time()
lr2.fit(X_reduced, y_train)
te3 = time.time()
t_stamp3 = te3 - ts3
t_stamp3

lr2.score(X_reduced_test,y_test)

rf2.score(X_reduced_test, y_test)

accuracy_score(rf2.predict(X_reduced), y_train)

accuracy_score(rf2.predict(X_reduced_test), y_test)

```
